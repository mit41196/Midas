# Extract Tweets from the User Page

```python
def fetch_tweets(twitter_handle):
    
    ''' --- Authorize User and fetch all tweets from the given tweeter-handle ---'''
    auth = OAuthHandler(consumer_api, consumer_api_secret)
    auth.set_access_token(access_token, access_token_secret)
    api = API(auth)
    
    ls = []
    image_count = []
    for status in Cursor(api.user_timeline, screen_name=twitter_handle, tweet_mode="extended").items():
        ls.append(status)
        media = status.entities.get('media',[])
        image_count.append(len(media))
    return ls, image_count
```

# Write Tweets in JSON File

```python
def write_tweets(tweets, file_name):
    
    '''--- Write all retrieved tweets in a JSON file ---'''
    with open(file_name, "w") as file:
        for tweet in tweets:
            json.dump(tweet._json, file)
            file.write("\n")
        file.close()
```

```python
def display_tabular(file_name):
    
    ''' --- Parse JSON file and read all tweets ---'''
    t = []
    with open(file_name, 'r') as file:
        for line in file.readlines():
            t.append(json.loads(line))
        file.close()
    return t
```
